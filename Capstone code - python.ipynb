```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

# -------------------------------
# Load the dataset
print(" Loading dataset from data/capstone.csv...")
df = pd.read_csv("/content/capstone.csv", low_memory=False, encoding="latin1")

# -------------------------------
# Plot selected feature distributions before preprocessing
print(" Plotting feature distributions...")
features_to_plot = [
    "Flow Duration",
    "Total Fwd Packets",
    "Total Backward Packets",
    "Fwd Packet Length Mean",
    "Bwd Packet Length Mean",
    "Flow IAT Mean",
    "Packet Length Variance"
]

for feature in features_to_plot:
    if feature not in df.columns:
        print(f" Skipping missing feature: {feature}")
        continue

    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    sns.histplot(data=df, x=feature, hue="Label", bins=50, kde=True, element="step", palette="Set2")
    plt.title(f"Histogram: {feature}")

    plt.subplot(1, 2, 2)
    sns.boxplot(data=df, x="Label", y=feature, palette="Set2")
    plt.title(f"Boxplot: {feature} by Label")

    plt.tight_layout()
    plt.show()

# -------------------------------
# Preprocess the data
print(" Cleaning data...")
df = df.dropna()
df = df.replace([np.inf, -np.inf], np.nan).dropna()
df['Label'] = df['Label'].apply(lambda x: 'BENIGN' if x == 'BENIGN' else 'ATTACK')

numeric_df = df.select_dtypes(include=[np.number])
numeric_df['Label'] = df['Label']

X = numeric_df.drop('Label', axis=1)
y = numeric_df['Label']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X = pd.DataFrame(X_scaled, columns=X.columns)

# -------------------------------
# Split the data
print(" Splitting data...")
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# -------------------------------
# Train and evaluate models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Naive Bayes": GaussianNB(),
    "Decision Tree": DecisionTreeClassifier()
}

plt.figure(figsize=(10, 6))

for name, model in models.items():
    print(f"\n Training and Evaluating: {name}")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else None

    print(" Classification Report:")
    print(classification_report(y_test, y_pred, digits=4))
    print(" Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    if y_proba is not None:
        y_test_bin = y_test.map({'BENIGN': 0, 'ATTACK': 1})
        auc = roc_auc_score(y_test_bin, y_proba)
        print(f" AUC-ROC Score: {auc:.4f}")
        fpr, tpr, _ = roc_curve(y_test_bin, y_proba)
        plt.plot(fpr, tpr, label=f"{name} (AUC = {auc:.2f})")

plt.title("ROC Curve Comparison")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
# Correlation Matrix
plt.figure(figsize=(16, 12))
corr_matrix = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr_matrix, annot=False, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Matrix", fontsize=16)
plt.tight_layout()
plt.show()

# ROC Curve for all models
plt.figure(figsize=(10, 6))

for name, model in models.items():
    if hasattr(model, 'predict_proba'):
        y_proba = model.predict_proba(X_test)[:, 1]
        y_test_bin = y_test.map({'BENIGN': 0, 'ATTACK': 1})
        fpr, tpr, _ = roc_curve(y_test_bin, y_proba)
        auc = roc_auc_score(y_test_bin, y_proba)
        plt.plot(fpr, tpr, label=f"{name} (AUC = {auc:.2f})")

plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
plt.title("ROC Curve Comparison", fontsize=14)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
from sklearn.metrics import ConfusionMatrixDisplay, roc_auc_score, roc_curve, confusion_matrix

def evaluate_and_plot(model_name, model, X_test, y_test):
    print(f"\n Evaluating: {model_name}")

    # Predictions and probabilities
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else None

    # --- Confusion Matrix ---
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["BENIGN", "ATTACK"])
    plt.figure(figsize=(6, 5))
    disp.plot(cmap="Blues")
    plt.title(f"Confusion Matrix: {model_name}")
    plt.grid(False)
    plt.show()

    # --- ROC Curve ---
    if y_proba is not None:
        y_test_bin = y_test.map({'BENIGN': 0, 'ATTACK': 1})
        fpr, tpr, _ = roc_curve(y_test_bin, y_proba)
        auc = roc_auc_score(y_test_bin, y_proba)

        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}", color="darkorange")
        plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
        plt.title(f"ROC Curve: {model_name}")
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.legend(loc="lower right")
        plt.grid(True)
        plt.tight_layout()
        plt.show()

evaluate_and_plot("Decision Tree", models["Decision Tree"], X_test, y_test)
evaluate_and_plot("Logistic Regression", models["Logistic Regression"], X_test, y_test)
evaluate_and_plot("Naive Bayes", models["Naive Bayes"], X_test, y_test)
```
